## 1. Variables & Expressions — Compound Interest Calculation

Calculating compound interest using the formula `A = P(1 + r/100)^n`.

```python
n = 10   # number of years
r = 5    # rate of interest (%)
p = 100  # principal amount

A = p * (1 + r / 100) ** n
print(f"Compound Amount: {A:.2f}")
```

## 13. Dictionaries

Mapping countries to their capitals, then updating a value.

```python
country_capitals = {
    "India": "Delhi",
    "China": "Beijing",
    "Japan": "Tokyo",
    "Qatar": "Doha",
    "France": "Marseilles"
}

# Update France's capital to Paris (two equivalent approaches)
country_capitals["France"] = "Paris"
country_capitals.update({"France": "Paris"})

print(country_capitals)
```

## 14. Tuples — Immutability

```python
tuple_1 = (1, 5, 6, 7, 8)
tuple_2 = (8, 9, 4)

print("Sum:", sum(tuple_1))
print("Length:", len(tuple_2))
print("Concatenated:", tuple_2 + tuple_1)

# Tuples are immutable - item assignment raises a TypeError
try:
    tuple_1[3] = 45
except TypeError as e:
    print("TypeError:", e)
```

## 16. Modular Function Design — Pythagorean Triplets

A function that finds all Pythagorean triplets `(a, b, c)` where no side exceeds a given natural number `N`.

```python
def pythagorean_triplets(N):
    """Return all Pythagorean triplets (a, b, c) with a <= b <= c <= N."""
    triplets = []
    for a in range(1, N + 1):
        for b in range(a, N + 1):
            for c in range(b, N + 1):
                if a * a + b * b == c * c:
                    triplets.append((a, b, c))
    return triplets

result = pythagorean_triplets(30)
print(f"Found {len(result)} triplets:")
print(result)
```
